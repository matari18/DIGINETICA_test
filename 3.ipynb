### Присвоение статуса доступности товара

Для определения доступности товара к покупке мы используем YML фиды:

- Есть глобальный фид, содержащий все товары из каталога.
- Есть региональные фиды для каждого региона, которые указывают доступность товара в конкретном регионе.

**Вопрос**: как присвоить статус доступности товару, если самого товара нет в региональном фиде?

### Ожидания:

- Описание процесса присвоения статуса товару в случае отсутствия региональных данных.

### Ответ:


Описание процесса присвоения статуса доступности товара при отсутствии его в региональном фиде.
Если товара нет в региональном фиде, это может означать, что его либо нет в наличии в данном регионе, либо он не продается там по каким-то другим причинам (например, логистическим или юридическим ограничениям). Чтобы определить статус доступности такого товара, можно предложить следующий процесс:

1. Определить наличие товара в глобальном фиде.
Если товар присутствует в глобальном фиде, это означает, что он теоретически доступен для продажи, хотя его региональная доступность может отличаться.

2. Проверить региональный фид.
Если товар отсутствует в региональном фиде, это указывает на то, что в данном регионе он не может быть приобретен по какой-либо причине (недоступен для покупки).

3. Присвоение статуса товара.
Если товар присутствует в глобальном фиде, но отсутствует в региональном фиде, ему можно присвоить один из следующих статусов:
  * "Недоступен в регионе".
Это явный статус, указывающий, что товар есть в каталоге, но по каким-то причинам его нельзя купить в конкретном регионе.
  * "Доступен на заказ". 
Если существует возможность заказать товар с доставкой из другого региона, этот статус может сигнализировать, что товар можно приобрести, но он будет доставлен не из локальных складов.
  * "Нет в наличии".
Этот статус может указывать, что товар временно отсутствует на региональном складе, но теоретически его можно заказать, когда появится в наличии.

4. Алгоритм обработки отсутствующих товаров в региональном фиде. 
Если товар есть в глобальном фиде и нет в региональном фиде, применяем логику присвоения статуса, исходя из политики компании:
Нет в наличии в регионе.
Либо, если товар можно доставить из другого региона: Доступен для доставки.
Если товар нет в глобальном фиде, его можно отметить как удаленный или архивный, что означает, что товар больше не продается.
